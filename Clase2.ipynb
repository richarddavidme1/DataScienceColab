```python
#Separacion de elementos

inmueble.Valor[0].split()
```

```python
#transformacion a una dataframe
inmueble.Valor.str.split()
```

```python
#transformacion a una dataframe y add nuevos datos
valor=inmueble.Valor.str.split(expand=True)
inmueble["Moneda"]=valor[0]
inmueble["Precio"]=valor[1]
inmueble.head(10)
```

```python
#tipo de dato se debe convertir objeto a entero
inmueble.info()
```

```python
#cambio de caracteres
inmueble["Precio"]=inmueble["Precio"].str.replace(".","",regex=True)
inmueble.head(10)
```

```python
#Cambio a valores flotantes
inmueble["Precio_Cambio"]=inmueble.Precio.astype("float")/1000000
inmueble.head(10)
```

```python
#comprobacion de valor float
inmueble.info()
```

```python
#Calculo de la media ,mediana,desviacion estandar ect
inmueble.describe()
```

```python
#Ajustar a dos cifras decimales

pd.set_option("display.precision",2)
pd.set_option("display.float_format",lambda x: "%.2f" %x)
inmueble.describe()
```

```python
#histogram
inmueble["Precio_Cambio"].plot.hist(bins=50)
```

```python
#Librerias de graficas y mejor manejo de datos
import matplotlib.pyplot as plt
import seaborn as sns

#Dibujo de graficas con dimension de (10,8)
plt.figure(figsize=(10,6))
#datos  para el histograma ,curva de distribucion,tipo
grafica=sns.histplot(data=inmueble,x="Precio_Cambio",kde=True,hue="Tipo")
#Titulo de la grafica
grafica.set_title("Distribucion de valores de inmuebles de Bogota")
#rangos min y max
plt.xlim((50,1000))
#guardar figura
plt.savefig("/content/drive/MyDrive/Colab Notebooks/Imag1",format="png")
#salida
plt.show()
```
